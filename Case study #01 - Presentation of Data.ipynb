# Case study #01 - Presentation of Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Read the dataset to the python environment.

```python
df = pd.read_csv('Downloads/cars_data.csv')
```

```python
df.head(10)
```

# 2. Check for the null values present in the dataset.

```python
df.isnull().sum()
```

# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
plt.figure(figsize=(5,5))
ax = df['Buyer Gender'].value_counts().plot(kind='bar');
ax.set_title('Bar graph of male vs female buyers participated in the Sales', fontsize=16)
ax.set(ylim=[4900, 5075])
sns.set_style("darkgrid")
plt.ylabel('No. of cars sold', fontsize=16)
plt.xlabel('Gender',fontsize=16)
plt.xticks(rotation='horizontal')
plt.show()
```
